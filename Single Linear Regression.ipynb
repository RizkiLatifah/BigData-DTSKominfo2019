```python
import pandas as pd
with pd.ExcelFile('E:\\Digitalent\karyawan.xls') as f:
    data = pd.read_excel(f, 'personalia1')
data
```

```python
import matplotlib.pyplot as plt
plt.scatter(data.Lama_kerja, data.Gaji_akhir,  color='blue')
plt.xlabel("Lama Kerja")
plt.ylabel("Gaji Akhir")
plt.show()
```

```python
import numpy as np
msk = np.random.rand(len(data)) < 0.8
train = data[msk]
test = data[~msk] 
```

```python
plt.scatter(train.Lama_kerja, train.Gaji_akhir,  color='blue')
plt.xlabel("Lama kerja")
plt.ylabel("Gaji akhir")
plt.show()
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['Lama_kerja']])
train_y = np.asanyarray(train[['Gaji_akhir']])
regr.fit (train_x, train_y)
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
plt.scatter(train.Lama_kerja, train.Gaji_akhir,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("Lama Kerja")
plt.ylabel("gaji Akhir")
```

```python
xnew = [[2.3]]
ynew = regr.predict(xnew)

print(ynew) 
```
